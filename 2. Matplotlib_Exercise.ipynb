# matplotlib

matplotlibを使ってグラフを描いてみます

### 直線のグラフ

```python
import math
import numpy as np
from matplotlib import pyplot

x = np.linspace(0,3,4)  # 0から3の間で等間隔に４個の値を生成します　→ xの値として(0, 1, 2, 3)を生成します
y = x + 1               # 直線の式

pyplot.plot(x,y)      # 直線を引きます
pyplot.show()              # グラフを描画します
```

### 曲線のグラフ

```python
import math
import numpy as np
from matplotlib import pyplot

pi = math.pi   #mathモジュールのπを利用

# x = np.linspace(0,3,4)  # 0から3の間で等間隔に４個の値を生成します　→ xの値として(0, 1, 2, 3)を生成します
x = np.linspace(0, 2*pi, 100)  #0から2πまでの範囲を100分割したnumpy配列
y = np.sin(x)

#凡例のためにlabelキーワードで凡例名をつける
pyplot.plot(x, y, label='sin')

#グラフタイトル
pyplot.title('Sin Graph')

#グラフの軸
pyplot.xlabel('X-Axis')
pyplot.ylabel('Y-Axis')

#グラフの凡例
pyplot.legend()

pyplot.plot(x,y, label='sin')      # 直線を引きます
pyplot.show()              # グラフを描画します
```

### 日本語化してみます

座標軸や凡例に日本語を使えるようにします

* 参考：[matplotlib のグラフに日本語を表示する方法（文字化け対応）](http://akiyoko.hatenablog.jp/entry/2017/04/11/080446)
1. [IPAexフォント - ダウンロード](http://ipafont.ipa.go.jp/node17#jp) → [IPAexフォント Ver.003.01](http://ipafont.ipa.go.jp/node26#jp)からipaexg00301.zip(3.92MB)をダウンロード
* 適当なフォルダーでunzipします
* `cp ~/Downloads/ipaexg00301/ipaexg.ttf ~/Library/Fonts/` でコピーします
* Jupyter notebookが停止していることを確認します
* `rm ~/.matplotlib/fontList*.cache` でmatplotlibのフォントキャッシュをクリアします
* jupyter notebookを実行し、下記を実行してみます

```python
import math
import numpy as np
from matplotlib import pyplot

font = {'family' : 'IPAexGothic', 'weight' : 'bold', 'size'   : 20}
#font = {'family' : 'AppleGothic', 'weight' : 'bold', 'size'   : 20}
pyplot.rc('font', **font)

pi = math.pi   #mathモジュールのπを利用

# x = np.linspace(0,3,4)  # 0から3の間で等間隔に４個の値を生成します　→ xの値として(0, 1, 2, 3)を生成します
x = np.linspace(0, 2*pi, 100)  #0から2πまでの範囲を100分割したnumpy配列
y = np.sin(x)

#グラフタイトル
pyplot.title('Sinグラフ')

#グラフの軸
pyplot.xlabel(u'X軸')
pyplot.ylabel(u'Y軸')

#グラフの凡例
# pyplot.legend()

#凡例のためにlabelキーワードで凡例名をつける
pyplot.plot(x, y, color='blue', label=u'sin値')

pyplot.legend(loc='upper right',
           bbox_to_anchor=(1, 0.5, 0.5, .100), 
           borderaxespad=0.,)

# pyplot.legend()
pyplot.show()

```
